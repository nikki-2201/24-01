```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 7, 6, 11]
plt.plot(x, y,color="hotpink",linestyle="--",marker="o",linewidth=3)  
plt.xlabel("X-axis ")
plt.ylabel("Y-axis ")
plt.title("Simple Line Plot")
plt.show()
```

```python
x=['jan','feb','mar','apr','may','june']
y=[1000,2000,2300,1700,2500,1200]
plt.plot(x,y,color="green",linestyle="dashed",marker="o")
plt.xlabel("months")
plt.ylabel("savings")
plt.title("Savings per month")
plt.show()
```

```python
x = ['A','B','C','D','E','F']
y = [2, 14, 7, 6, 11, 15]
plt.bar(x, y,color=["hotpink","skyblue"],edgecolor="black",width=0.75)  
plt.xlabel("X-axis ")
plt.ylabel("Y-axis ")
plt.title("Simple Bar Plot")
plt.show()
```

```python
ages=[20, 25, 31, 40, 17, 22, 39, 56, 28, 33,23, 34, 18, 45, 29, 61]
b=[10,20,30,40,50,60]
plt.hist(ages,bins=b,color="skyblue",edgecolor="black")
plt.xlabel("X-axis ")
plt.ylabel("Y-axis ")
plt.title("Simple Hist Plot")
plt.grid(True)
plt.show()
```

```python
x = ['A','B','C','D','E','F']
y = [2, 4, 7, 6, 5, 6]
colors = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99', 'lightpink', 'skyblue']

plt.pie(y, labels=x, autopct='%.1f%%', colors=colors,startangle=90)
plt.legend()
plt.show()
```

```python
x = [20, 25, 31, 40, 17,12]
y = [2, 4, 7, 6, 5, 6]
plt.scatter(x,y,color="green",s=100,edgecolor="black")
plt.xlabel("X-axis ")
plt.ylabel("Y-axis ")
plt.show()
```
